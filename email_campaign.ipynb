```python
# Add columns for word and character counts
data_temp['body_word_count'] = data_temp['body'].apply(lambda x: len(x.split()))
data_temp['body_char_count'] = data_temp['body'].apply(lambda x: len(x))
data_temp['subject_word_count'] = data_temp['subject'].apply(lambda x: len(x.split()))
data_temp['subject_char_count'] = data_temp['subject'].apply(lambda x: len(x))

# Calculate average word length
data_temp['body_avg_word_length'] = data_temp['body_char_count'] / data_temp['body_word_count']
data_temp['subject_avg_word_length'] = data_temp['subject_char_count'] / data_temp['subject_word_count']

print(data_temp[['body_word_count', 'body_char_count', 'body_avg_word_length', 'subject_word_count', 'subject_char_count', 'subject_avg_word_length']].head())
```

```python
from collections import Counter
import re

def preprocess_text(text):
    # Remove non-alphanumeric characters and convert to lower case
    text = re.sub(r'\W+', ' ', text).lower()
    return text

# Combine all bodies and subjects into single texts
all_body_text = ' '.join(data_temp['body'].apply(preprocess_text))
all_subject_text = ' '.join(data_temp['subject'].apply(preprocess_text))

# Calculate word frequencies
body_word_freq = Counter(all_body_text.split())
subject_word_freq = Counter(all_subject_text.split())

print("Most common words in body:")
print(body_word_freq.most_common(10))

print("Most common words in subject:")
print(subject_word_freq.most_common(10))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Preprocess text data
data_temp['processed_body'] = data_temp['body'].apply(preprocess_text)
data_temp['processed_subject'] = data_temp['subject'].apply(preprocess_text)

# Vectorize text data
vectorizer_body = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
body_dtm = vectorizer_body.fit_transform(data_temp['processed_body'])

vectorizer_subject = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
subject_dtm = vectorizer_subject.fit_transform(data_temp['processed_subject'])

# Fit LDA model
lda_body = LatentDirichletAllocation(n_components=5, random_state=42)
lda_body.fit(body_dtm)

lda_subject = LatentDirichletAllocation(n_components=5, random_state=42)
lda_subject.fit(subject_dtm)

# Display topics
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1] if i < len(feature_names)]))

no_top_words = 10
print("Topics in body:")
display_topics(lda_body, vectorizer_body.get_feature_names_out(), no_top_words)

print("Topics in subject:")
display_topics(lda_subject, vectorizer_subject.get_feature_names_out(), no_top_words)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
text_features = ['body_word_count', 'body_char_count', 'body_avg_word_length', 'body_sentiment', 'subject_word_count', 'subject_char_count', 'subject_avg_word_length', 'subject_sentiment']
correlation_matrix = data_temp[text_features + ['opened', 'responded']].corr()
print(correlation_matrix)
# Plot correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Text Features and Engagement Metrics')
plt.show()
```

* There is a noticeable negative correlation, suggesting that as the number of words in the subject increases, the average word length decreases. This could imply that longer subjects tend to use shorter words.

* There is a moderate positive correlation, indicating that longer average word lengths in the body might slightly correlate with positive sentiment.

* Subjects with longer avergae word would receive few responses

* Many correlations are weak, indicating little to no linear relationship between those pairs of variables. For instance, opened has weak correlations with most other variables, meaning the factors considered do not strongly influence whether an email is opened.
